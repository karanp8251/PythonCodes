```python
countries = ['India','Australia','USA']
for index, country in enumerate(countries):
    place = str(index)
    print('Index:' +place + ", Value: " +country)

    index = countries.index('India')
    print("\n Index of India is: " ,index)
```

```python
dict = {"Sedan": 1500, "SUV": 2000, "Pickup": 2500}
d = [i.lower() for i in dict if dict[i]<5000]
print(d)
```

```python
sentence = 'In 1984 there were 13 instances of a protest with over 1000 people attending'
words = sentence.split()
result = [number for number in words if not number.isalpha()]
print(result)
```

```python
list_a = [1,2,3,4]
list_b = [2,3,4,5]
common = [a for a in list_a if a in list_b]
print(common)
```
